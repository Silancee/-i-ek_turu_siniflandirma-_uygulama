```python
import torch
import clip
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files
from gtts import gTTS
from IPython.display import Audio, display


device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)
print("âœ… CLIP modeli yÃ¼klendi")


flower_classes = [
    "rose", "tulip", "sunflower", "daisy", "lily",
    "orchid", "lavender", "carnation", "violet",
    "hyacinth", "jasmine", "peony", "begonia",
    "hydrangea", "daffodil", "cactus",

    "lotus", "hibiscus", "magnolia", "camellia",
    "poppy", "iris", "anemone", "freesia",
    "chrysanthemum", "geranium", "petunia",
    "zinnia", "gardenia", "azalea", "foxglove",

    "clematis", "bougainvillea", "oleander",
    "wisteria", "morning glory"
]


flower_info = {
    "rose": "GÃ¼l, hoÅŸ kokusu ve estetik gÃ¶rÃ¼nÃ¼mÃ¼yle bilinen Ã§ok yÄ±llÄ±k bir sÃ¼s bitkisidir.",
    "tulip": "Lale, ilkbaharda aÃ§an ve soÄŸanlÄ± yapÄ±ya sahip bir Ã§iÃ§ektir.",
    "sunflower": "AyÃ§iÃ§eÄŸi, gÃ¼neÅŸe yÃ¶nelme Ã¶zelliÄŸiyle bilinen bÃ¼yÃ¼k sarÄ± Ã§iÃ§eklere sahiptir.",
    "daisy": "Papatya, beyaz yapraklarÄ± ve sarÄ± merkeziyle tanÄ±nÄ±r.",
    "lily": "Zambak, bÃ¼yÃ¼k ve hoÅŸ kokulu Ã§iÃ§eklere sahip zarif bir bitkidir.",
    "orchid": "Orkide, tropikal bÃ¶lgelerde yetiÅŸen ve Ã§ok sayÄ±da tÃ¼re sahip bir Ã§iÃ§ektir.",
    "lavender": "Lavanta, mor renkli Ã§iÃ§ekleri ve rahatlatÄ±cÄ± kokusuyla bilinir.",
    "carnation": "Karanfil, uzun Ã¶mÃ¼rlÃ¼ ve hoÅŸ kokulu bir sÃ¼s bitkisidir.",
    "violet": "MenekÅŸe, kÃ¼Ã§Ã¼k yapÄ±lÄ± ve mor tonlarÄ±nda Ã§iÃ§eklere sahiptir.",
    "hyacinth": "SÃ¼mbÃ¼l, yoÄŸun kokulu ve salkÄ±m ÅŸeklinde Ã§iÃ§eklere sahiptir.",
    "jasmine": "Yasemin, beyaz Ã§iÃ§ekleri ve gÃ¼Ã§lÃ¼ kokusuyla tanÄ±nÄ±r.",
    "peony": "ÅžakayÄ±k, bÃ¼yÃ¼k ve katmanlÄ± yapraklara sahip gÃ¶steriÅŸli bir Ã§iÃ§ektir.",
    "begonia": "Begonya, iÃ§ ve dÄ±ÅŸ mekÃ¢nda yetiÅŸtirilebilen sÃ¼s bitkisidir.",
    "hydrangea": "Ortanca, toprak pHâ€™Ä±na gÃ¶re renk deÄŸiÅŸtirebilen Ã§iÃ§eklere sahiptir.",
    "daffodil": "Nergis, ilkbaharÄ±n habercisi olan sarÄ± veya beyaz Ã§iÃ§eklere sahiptir.",
    "cactus": "KaktÃ¼s, kurak ortamlara dayanÄ±klÄ± ve dikenli yapÄ±ya sahip bir bitkidir.",

    "lotus": "Lotus, su Ã¼zerinde yetiÅŸen ve saflÄ±ÄŸÄ±n simgesi olarak bilinen bir Ã§iÃ§ektir.",
    "hibiscus": "Hibiskus, bÃ¼yÃ¼k ve renkli Ã§iÃ§ekleriyle bilinen tropikal bir bitkidir.",
    "magnolia": "Manolya, bÃ¼yÃ¼k beyaz veya pembe Ã§iÃ§ekleri olan bir aÄŸaÃ§ tÃ¼rÃ¼dÃ¼r.",
    "camellia": "Kamelya, parlak yapraklÄ± ve gÃ¶steriÅŸli Ã§iÃ§eklere sahiptir.",
    "poppy": "Gelincik, kÄ±rmÄ±zÄ± renkli ve ince yapraklÄ± bir kÄ±r Ã§iÃ§eÄŸidir.",
    "iris": "SÃ¼sen, farklÄ± renklerde Ã§iÃ§eklere sahip Ã§ok yÄ±llÄ±k bir bitkidir.",
    "anemone": "Anemon, narin yapÄ±lÄ± ve renkli Ã§iÃ§eklere sahip bir bitkidir.",
    "freesia": "Frezya, hoÅŸ kokulu ve renkli Ã§iÃ§ekleriyle bilinir.",
    "chrysanthemum": "KasÄ±mpatÄ±, sonbaharda aÃ§an dayanÄ±klÄ± bir sÃ¼s bitkisidir.",
    "geranium": "Sardunya, balkon ve bahÃ§elerde sÄ±kÃ§a yetiÅŸtirilen bir bitkidir.",
    "petunia": "Petunya, sarkÄ±cÄ± yapÄ±sÄ± ve canlÄ± renkleriyle bilinir.",
    "zinnia": "Zinya, yaz aylarÄ±nda aÃ§an dayanÄ±klÄ± bir Ã§iÃ§ektir.",
    "gardenia": "Gardenya, beyaz Ã§iÃ§ekleri ve yoÄŸun kokusuyla tanÄ±nÄ±r.",
    "azalea": "AÃ§elya, Ã§alÄ± formunda ve renkli Ã§iÃ§eklere sahip bir bitkidir.",
    "foxglove": "YÃ¼ksÃ¼k otu, Ã§an ÅŸeklindeki Ã§iÃ§ekleriyle bilinir.",

    "clematis": "Akasma, sarmaÅŸÄ±k formunda ve bÃ¼yÃ¼k Ã§iÃ§eklere sahip bir bitkidir.",
    "bougainvillea": "Begonvil, canlÄ± renkli yapraklarÄ±yla bilinen bir sarmaÅŸÄ±ktÄ±r.",
    "oleander": "Zakkum, sÄ±cak iklimlerde yetiÅŸen dayanÄ±klÄ± bir Ã§alÄ±dÄ±r.",
    "wisteria": "Mor salkÄ±m, sarkÄ±k mor Ã§iÃ§ekleriyle bilinen bir sarmaÅŸÄ±ktÄ±r.",
    "morning glory": "GÃ¼ndÃ¼z sefasÄ±, sabah saatlerinde aÃ§an renkli Ã§iÃ§eklere sahiptir."
}


text_inputs = torch.cat([
    clip.tokenize(f"a photo of a {c} flower") for c in flower_classes
]).to(device)


print("ðŸ“· Bir Ã§iÃ§ek fotoÄŸrafÄ± yÃ¼kleyin")
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

image = Image.open(image_path).convert("RGB")
image_input = preprocess(image).unsqueeze(0).to(device)


with torch.no_grad():
    image_features = model.encode_image(image_input)
    text_features = model.encode_text(text_inputs)

    image_features /= image_features.norm(dim=-1, keepdim=True)
    text_features /= text_features.norm(dim=-1, keepdim=True)

    similarity = (100.0 * image_features @ text_features.T).softmax(dim=-1)

values, indices = similarity[0].topk(5)


plt.imshow(image)
plt.axis("off")
plt.show()

print("ðŸŒ¸ Tahmin SonuÃ§larÄ± (Top-5):\n")
for value, index in zip(values, indices):
    print(f"{flower_classes[index]}  -->  %{value.item()*100:.2f}")


best_index = indices[0].item()
best_flower = flower_classes[best_index]
best_confidence = values[0].item() * 100
info_text = flower_info.get(best_flower, "Bu Ã§iÃ§ek hakkÄ±nda bilgi bulunmamaktadÄ±r.")

print("\nðŸŒ¼ En OlasÄ± Ã‡iÃ§ek TÃ¼rÃ¼:")
print(f"{best_flower}  -->  %{best_confidence:.2f}")
print("\nðŸ“˜ Ã‡iÃ§ek HakkÄ±nda Bilgi:")
print(info_text)


speech_text = f"Tahmin edilen Ã§iÃ§ek tÃ¼rÃ¼ {best_flower}. {info_text}"
tts = gTTS(text=speech_text, lang="tr")
tts.save("flower_info.mp3")

print("\nðŸ”Š Sesli anlatÄ±m:")
display(Audio("flower_info.mp3", autoplay=True))
```
